# Join the respective table to integrate the fact sales table

```python
from pyspark.sql.functions import current_timestamp

query = """
SELECT
    sd.order_number,
    pr.product_key,
    cu.customer_key,
    sd.order_date,
    sd.ship_date,
    sd.due_date,
    sd.sales_amount,
    sd.quantity,
    sd.price
FROM dev_project.silver.crm_sales_details sd
LEFT JOIN dev_project.gold.gold_dim_products pr
    ON sd.product_number = pr.product_number
LEFT JOIN dev_project.gold.gold_dim_customers cu
    ON sd.customer_id = cu.customer_id;
"""
df = spark.sql(query)
```

```python
df = df.withColumn("_ingest_time", current_timestamp())
```

```python
df.display()
```

# write fact sales toble into Gold

```python
df.write.mode("overwrite").saveAsTable("dev_project.gold.gold_fact_sales")
```
