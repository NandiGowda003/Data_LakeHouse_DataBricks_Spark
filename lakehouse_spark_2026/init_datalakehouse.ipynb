# Select Catalog

```python
%sql
USE CATALOG workspace
```

# Create lakehouse schema

```python
%sql
CREATE SCHEMA IF NOT EXISTS bronze
COMMENT 'Bronze layer: raw data ingested';
    
CREATE SCHEMA IF NOT EXISTS silver
COMMENT 'Silver layer: clean data';

CREATE SCHEMA IF NOT EXISTS gold
COMMENT 'Gold layer: aggregated data';
```

# Create Volume

```python
%sql
CREATE VOLUME IF NOT EXISTS workspace.bronze.raw_sources
COMMENT 'Volume for RAW sources Files'
```
